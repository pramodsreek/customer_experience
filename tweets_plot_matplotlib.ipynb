# Plot time series using matplotlib

````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# file name and path for the tweets.
user_handle = "business"
path = "data/"
file_extension = ".csv"
````

````python
file_user = path + user_handle + file_extension
df_user = pd.read_csv(file_user, index_col=5, parse_dates=["date"])
````

````python
df_user.head(2)
````

````python
time_likes_user = pd.Series(data=df_user['likes'].values, index=df_user['date'])
time_likes_user.plot(figsize=(16, 4), label='likes', legend=True)
````

````python
time_sentiment_user = pd.Series(data=df_user['sentiment'].values, index=df_user['date'])
time_sentiment_user.plot(figsize=(16, 4), label='sentiment', legend=True)
````

````python
another_handle = "economics"
file_another = path + another_handle + file_extension
````

````python
df_another = pd.read_csv(file_another, index_col=5, parse_dates=["date"])
````

````python
plt.figure(figsize=(16,5), dpi=100)
plt.plot(df_another['date'], df_another['sentiment'].values, color='tab:red')
plt.gca().set(title='just right', xlabel='date', ylabel='value')
plt.show()
````
